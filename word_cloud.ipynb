```python
from janome.tokenizer import Tokenizer
tokenizer = Tokenizer()

tokens = tokenizer.tokenize(text)
nouns = []
verbs = []
adjectives = []
for token in tokens:
  if token.part_of_speech.split(",")[0] == '名詞':
    nouns.append(token.surface)
  elif token.part_of_speech.split(",")[0] == '動詞':
    verbs.append(token.surface)
  elif token.part_of_speech.split(",")[0] in ['形容詞','副詞']:
    adjectives.append(token.surface)
```

```python
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
  if word in nouns:
    return "rgb(200, 100, 100)" # red for nouns
  elif word in verbs:
    return "rgb(100, 200, 100)" # green for verbs
  elif word in adjectives:
    return "rgb(100, 100, 200)" # blue for adjectives
  else:
    return "rgb(200, 200, 200)" # black for the rest
```

```python
analyzer = Analyzer(token_filters=[TokenCountFilter()])
freq = dict(filter(lambda p: len(p[0]) >= 2, analyzer.analyze(text)))
print(freq)
```

```python
wordcloud = WordCloud(
font_path="/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"
,
width=800,
height=600,
background_color="white",
color_func=color_func,
prefer_horizontal=1.0
).generate_from_frequencies(freq)
wordcloud.to_image()
```

```python
!pip install wordcloud Janome
!apt-get install -yq fonts-noto-cjk
from wordcloud import WordCloud
from janome.tokenfilter import POSKeepFilter, TokenCountFilter
from janome.analyzer import Analyzer
import requests
from bs4 import BeautifulSoup
```

```python
def get_text(url):
  text_in_webpage = []
  r = requests.get(url)
  c = r.content
  soup = BeautifulSoup(c, "html.parser")
  p_tag_content = soup.find_all("p")
  temp = []
  for con in p_tag_content:
    out = con.text
    temp.append(out)
    text_in_webpage = ''.join(temp)
  return text_in_webpage
```

```python
text = get_text("https://www.tachibana-u.ac.jp/news/2022/08/post-1178.html")
analyzer = Analyzer(token_filters=[POSKeepFilter('名詞'), TokenCountFilter(sorted=True)])
freq = dict(filter(lambda p: len(p[0]) >= 1, analyzer.analyze(text)))
set1 = set(freq.keys())
print(set1)
```

```python
def compute_similarity(set1, set2):
  total_similarity = 0
  count = 0
  for word1 in set1:
    for word2 in set2:
      # 単語がモデルの語彙に存在するか確認
      if word1 in model.key_to_index and word2 in model.key_to_index:
        similarity = model.similarity(word1, word2)
        total_similarity += similarity
        count += 1
      # 平均類似度を計算
      if count != 0:
        average_similarity = total_similarity / count
      else:
        average_similarity = 0 # 単語がモデルの語彙に存在しない場合
  return average_similarity
```
